```python
with open("/content/sample_data/sample.jpg", "rb") as f:
    image_bytes = f.read()
image_b64 = base64.b64encode(image_bytes).decode("utf-8")
image_b64
```

```python
response = client.interactions.create(
    model="gemini-3.5-flash",
    input=[

        {
            "type": "image",
            "data": image_b64,
            "mime_type": "image/jpeg"
        },

        {
            "type": "text",
            "text": "define the image in 3 lines",

        }

    ]
)
print(response.output_text)
```

```python
response
```

```python
with open("/content/sample_data/pdf_file.pdf", "rb") as f:
    pdf_bytes = f.read()

pdf_b64 = base64.b64encode(pdf_bytes).decode("utf-8")

interaction = client.interactions.create(
    model="gemini-3.5-flash",
    input=[
        {
            "type": "text",
            "text": "Summarize this PDF in 3 lines."
        },
        {
            "type": "document",
            "data": pdf_b64,
            "mime_type": "application/pdf"
        }
    ]
)

print(interaction.output_text)
```
